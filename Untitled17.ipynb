```python
"""Download medical imaging datasets from Kaggle.

Before running:
    pip install kagglehub

Make sure your Kaggle API credentials are configured. By default, datasets are
stored under E:\\Download\\Pulmo Vision\\datasets. You can override this with the
PULMO_VISION_DATASETS environment variable.
"""

from __future__ import annotations

import os
import shutil
from pathlib import Path

import kagglehub


BASE_DIR = Path(
    os.environ.get(
        "PULMO_VISION_DATASETS",
        r"datasetsxctm",
    )
)


DATASETS = {
    "ct": [
        ("mehradaria/covid19-lung-ct-scans", "covid19_lung_ct_scans"),
        ("plameneduardo/sarscov2-ctscan-dataset", "sarscov2_ctscan"),
        ("kmader/finding-lungs-in-ct-data", "finding_lungs_ct"),
        ("andrewmvd/mosmed-covid19-ct-scans", "mosmed_ct_scans"),
        (
            "programmer3/lung-ct-and-histopathological-images-dataset",
            "lung_ct_histopath",
        ),
    ],
    "xray": [

        ("tawsifurrahman/covid19-radiography-database", "covid19_radiography"),
        ("yazanqiblawey/sars-mers-xray-images-dataset", "sars_mers_xray"),
        ("subhankarsen/novel-covid19-chestxray-repository", "novel_covid_xray"),
        ("nikhilpandey360/chest-xray-masks-and-labels", "xray_masks_labels"),
    ],
    "mri": [
        ("masoudnickparvar/brain-tumor-mri-dataset", "brain_tumor_mri"),
        ("tourist55/alzheimers-dataset-4-class-of-images", "alzheimers_mri"),
        (
            "navoneel/brain-mri-images-for-brain-tumor-detection",
            "brain_mri_detection",
        ),
        ("andrewmvd/multiple-sclerosis-mri-dataset", "multiple_sclerosis_mri"),
        ("sartajbhuvaji/brain-tumor-classification-mri", "brain_tumor_classification"),
    ],
}


def download_and_store(dataset_name: str, destination: Path) -> None:
    """Download a Kaggle dataset and copy it into the destination folder."""
    print(f"\nDownloading: {dataset_name}")

    try:
        downloaded_path = Path(kagglehub.dataset_download(dataset_name))

        if destination.exists():
            shutil.rmtree(destination)

        destination.parent.mkdir(parents=True, exist_ok=True)
        shutil.copytree(downloaded_path, destination)

        print(f"Saved: {destination}")
    except Exception as exc:
        print(f"Failed: {dataset_name}")
        print(exc)


def main() -> None:
    BASE_DIR.mkdir(parents=True, exist_ok=True)

    for category, dataset_list in DATASETS.items():
        category_dir = BASE_DIR / category
        category_dir.mkdir(parents=True, exist_ok=True)

        for dataset_name, folder_name in dataset_list:
            destination = category_dir / folder_name
            download_and_store(dataset_name, destination)

    print("\nAll datasets downloaded.")


if __name__ == "__main__":
    main()
```

```python
from __future__ import annotations

import csv
import shutil
import os
from dataclasses import dataclass
from enum import Enum
from pathlib import Path
from typing import Callable

from PIL import Image, ImageOps
from tqdm.auto import tqdm


class ImageMode(str, Enum):
    RGB = "RGB"
    L = "L"


@dataclass(frozen=True)
class PreprocessConfig:
    image_size: tuple[int, int] = (224, 224)
    output_mode: ImageMode = ImageMode.L
    overwrite: bool = True
    copy_non_images: bool = False

def iter_files(path: Path):
    for root, _, files in os.walk(path):
        for file in files:
            yield Path(root) / file

def is_image_file(path: Path) -> bool:
    return path.suffix.lower() in [".jpg", ".jpeg", ".png", ".gif", ".bmp", ".tiff", ".webp"]

def prepare_output_dir(output_dir: Path, overwrite: bool) -> None:
    if overwrite and output_dir.exists():
        shutil.rmtree(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)


def load_image(path: Path, mode: str) -> Image.Image:
    with Image.open(path) as image:
        image = ImageOps.exif_transpose(image)
        return image.convert(mode)


def resize_image(image: Image.Image, image_size: tuple[int, int]) -> Image.Image:
    return ImageOps.fit(
        image,
        image_size,
        method=Image.Resampling.LANCZOS,
        centering=(0.5, 0.5),
    )


def preprocess_ct_image(image: Image.Image, image_size: tuple[int, int]) -> Image.Image:
    image = ImageOps.grayscale(image)
    image = ImageOps.autocontrast(image, cutoff=1)
    return resize_image(image, image_size)


def preprocess_xray_image(image: Image.Image, image_size: tuple[int, int]) -> Image.Image:
    image = ImageOps.grayscale(image)
    image = ImageOps.equalize(image)
    image = ImageOps.autocontrast(image, cutoff=1)
    return resize_image(image, image_size)


def preprocess_mri_image(image: Image.Image, image_size: tuple[int, int]) -> Image.Image:
    image = ImageOps.grayscale(image)
    image = ImageOps.autocontrast(image, cutoff=2)
    return resize_image(image, image_size)


def output_image_path(source_file: Path, source_dir: Path, output_dir: Path) -> Path:
    relative_path = source_file.relative_to(source_dir)
    return (output_dir / relative_path).with_suffix(".png")


def preprocess_dataset(
    source_dir: str | Path,
    output_dir: str | Path,
    image_processor: Callable[[Image.Image, tuple[int, int]], Image.Image],
    config: PreprocessConfig | None = None,
) -> list[dict[str, str]]:
    """Preprocess every image in a dataset folder and write a CSV manifest."""
    config = config or PreprocessConfig()
    source_dir = Path(source_dir)
    output_dir = Path(output_dir)

    if not source_dir.exists():
        raise FileNotFoundError(f"Source dataset folder does not exist: {source_dir}")

    prepare_output_dir(output_dir, overwrite=config.overwrite)

    records: list[dict[str, str]] = []
    all_files = list(iter_files(source_dir))

    for source_file in tqdm(all_files, desc=f"Processing {source_dir.name}"):
        if not is_image_file(source_file):
            if config.copy_non_images:
                target_file = output_dir / source_file.relative_to(source_dir)
                target_file.parent.mkdir(parents=True, exist_ok=True)
                shutil.copy2(source_file, target_file)
            continue

        target_file = output_image_path(source_file, source_dir, output_dir)
        target_file.parent.mkdir(parents=True, exist_ok=True)

        try:
            image = load_image(source_file, config.output_mode)
            processed_image = image_processor(image, config.image_size)
            processed_image.save(target_file, format="PNG", optimize=True)

            records.append(
                {
                    "source_path": str(source_file),
                    "processed_path": str(target_file),
                    "dataset_group": source_file.parent.name,
                    "status": "ok",
                }
            )
        except Exception as exc:
            records.append(
                {
                    "source_path": str(source_file),
                    "processed_path": "",
                    "dataset_group": source_file.parent.name,
                    "status": f"failed: {exc}",
                }
            )

    write_manifest(output_dir / "preprocess_manifest.csv", records)
    return records


def write_manifest(manifest_path: Path, records: list[dict[str, str]]) -> None:
    manifest_path.parent.mkdir(parents=True, exist_ok=True)
    fieldnames = ["source_path", "processed_path", "dataset_group", "status"]

    with manifest_path.open("w", newline="", encoding="utf-8") as csv_file:
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(records)


def preprocess_ct_dataset(
    source_dir: str | Path,
    output_dir: str | Path,
    image_size: tuple[int, int] = (224, 224),
    overwrite: bool = True,
) -> list[dict[str, str]]:
    config = PreprocessConfig(image_size=image_size, overwrite=overwrite)
    return preprocess_dataset(source_dir, output_dir, preprocess_ct_image, config)


def preprocess_xray_dataset(
    source_dir: str | Path,
    output_dir: str | Path,
    image_size: tuple[int, int] = (224, 224),
    overwrite: bool = True,
) -> list[dict[str, str]]:
    config = PreprocessConfig(image_size=image_size, overwrite=overwrite)
    return preprocess_dataset(source_dir, output_dir, preprocess_xray_image, config)


def preprocess_mri_dataset(
    source_dir: str | Path,
    output_dir: str | Path,
    image_size: tuple[int, int] = (224, 224),
    overwrite: bool = True,
) -> list[dict[str, str]]:
    config = PreprocessConfig(image_size=image_size, overwrite=overwrite)
    return preprocess_dataset(source_dir, output_dir, preprocess_mri_image, config)


if __name__ == "__main__":
    base_dir = Path(os.environ.get("PULMO_VISION_DATASETS", "datasetsxctm"))
    processed_dir = base_dir / "processed"

    preprocess_ct_dataset(base_dir / "ct", processed_dir / "ct")
    preprocess_xray_dataset(base_dir / "xray", processed_dir / "xray")
    preprocess_mri_dataset(base_dir / "mri", processed_dir / "mri")
```

```python
from __future__ import annotations
import csv
import shutil
import random
import os
from dataclasses import dataclass
from pathlib import Path
from typing import Iterable, Callable, Dict, List

@dataclass(frozen=True)
class SplitConfig:
    train_ratio: float = 0.7
    val_ratio: float = 0.15
    test_ratio: float = 0.15
    seed: int = 42
    overwrite: bool = True
    copy_files: bool = True

def validate_split_ratios(config: SplitConfig):
    if not (0.99 <= config.train_ratio + config.val_ratio + config.test_ratio <= 1.01):
        raise ValueError("Split ratios must sum to 1.0")

def find_class_dirs(source_dir: Path) -> List[Path]:
    # Returns subdirectories that contain images
    return [d for d in source_dir.iterdir() if d.is_dir()]

def class_name_for(class_dir: Path, source_dir: Path) -> str:
    return class_dir.name

def collect_images(class_dir: Path) -> List[Path]:
    extensions = (".jpg", ".jpeg", ".png", ".bmp")
    return [f for f in class_dir.rglob("*") if f.suffix.lower() in extensions]

def split_items(items: List[Path], config: SplitConfig) -> Dict[str, List[Path]]:
    random.seed(config.seed)
    random.shuffle(items)
    n = len(items)
    train_idx = int(n * config.train_ratio)
    val_idx = train_idx + int(n * config.val_ratio)
    return {
        "train": items[:train_idx],
        "val": items[train_idx:val_idx],
        "test": items[val_idx:]
    }

def copy_or_reference_file(source_file: Path, source_dir: Path, output_dir: Path, split_name: str, class_name: str, copy_files: bool) -> Path:
    target_file = output_dir / split_name / class_name / source_file.name
    if copy_files:
        target_file.parent.mkdir(parents=True, exist_ok=True)
        shutil.copy2(source_file, target_file)
    return target_file

def split_dataset(source_dir: str | Path, output_dir: str | Path, config: SplitConfig | None = None) -> list[dict[str, str]]:
    config = config or SplitConfig()
    validate_split_ratios(config)
    source_dir, output_dir = Path(source_dir), Path(output_dir)
    if not source_dir.exists(): raise FileNotFoundError(f"Source missing: {source_dir}")
    if config.overwrite and output_dir.exists(): shutil.rmtree(output_dir)

    records = []
    for class_dir in find_class_dirs(source_dir):
        class_name = class_name_for(class_dir, source_dir)
        split_map = split_items(collect_images(class_dir), config)
        for split_name, files in split_map.items():
            for source_file in files:
                target_file = copy_or_reference_file(source_file, source_dir, output_dir, split_name, class_name, config.copy_files)
                records.append({"source_path": str(source_file), "split_path": str(target_file), "dataset_group": class_name, "split": split_name})
    return records

def split_ct_dataset(source_dir: str | Path, output_dir: str | Path) -> list[dict[str, str]]:
    return split_dataset(source_dir, output_dir, SplitConfig())

def split_xray_dataset(source_dir: str | Path, output_dir: str | Path) -> list[dict[str, str]]:
    return split_dataset(source_dir, output_dir, SplitConfig())

def split_mri_dataset(source_dir: str | Path, output_dir: str | Path) -> list[dict[str, str]]:
    return split_dataset(source_dir, output_dir, SplitConfig())

if __name__ == "__main__":
    base_dir = Path(os.environ.get("PULMO_VISION_DATASETS", "datasetsxctm")) / "processed"
    split_dir = base_dir.parent / "splits"

    if (base_dir / "ct").exists(): split_ct_dataset(base_dir / "ct", split_dir / "ct")
    if (base_dir / "xray").exists(): split_xray_dataset(base_dir / "xray", split_dir / "xray")
    if (base_dir / "mri").exists(): split_mri_dataset(base_dir / "mri", split_dir / "mri")
    print("Splitting complete.")
```

```python
"""Load trained CT, X-ray, and MRI models.

Supported model formats:
    - TensorFlow/Keras: .keras, .h5, .hdf5, or SavedModel directories
    - PyTorch: .pt, .pth, .ckpt

The imports are lazy, so this file can be imported even when TensorFlow or
PyTorch is not installed yet. Install the framework that matches your saved
model files before loading them.
"""

from __future__ import annotations

import os
from dataclasses import dataclass
from pathlib import Path
from typing import Any, Literal


Framework = Literal["auto", "keras", "torch"]


KERAS_EXTENSIONS = {".h5", ".hdf5", ".keras"}
TORCH_EXTENSIONS = {".ckpt", ".pt", ".pth"}


@dataclass(frozen=True)
class ModelLoadConfig:
    model_path: str | Path
    framework: Framework = "auto"
    device: str = "cpu"
    compile_keras: bool = False
    eval_torch: bool = True


def ensure_model_exists(model_path: Path) -> None:
    if not model_path.exists():
        raise FileNotFoundError(
            f"Model file or directory does not exist: {model_path}. "
            "Please ensure you have trained the models or uploaded them to this path."
        )


def infer_framework(model_path: Path) -> Literal["keras", "torch"]:
    suffix = model_path.suffix.lower()

    if model_path.is_dir() or suffix in KERAS_EXTENSIONS:
        return "keras"

    if suffix in TORCH_EXTENSIONS:
        return "torch"

    raise ValueError(
        "Could not infer model framework from path. "
        "Use framework='keras' or framework='torch'."
    )


def load_keras_model(model_path: str | Path, compile_model: bool = False) -> Any:
    """Load a TensorFlow/Keras model."""
    model_path = Path(model_path)
    ensure_model_exists(model_path)

    try:
        from tensorflow import keras
    except ImportError as exc:
        raise ImportError(
            "TensorFlow is required to load Keras models. Install it with: "
            "pip install tensorflow"
        ) from exc

    return keras.models.load_model(model_path, compile=compile_model)


def load_torch_model(
    model_path: str | Path,
    device: str = "cpu",
    eval_mode: bool = True,
) -> Any:
    """Load a PyTorch model or checkpoint."""
    model_path = Path(model_path)
    ensure_model_exists(model_path)

    try:
        import torch
    except ImportError as exc:
        raise ImportError(
            "PyTorch is required to load Torch models. Install it with: "
            "pip install torch"
        ) from exc

    model = torch.load(model_path, map_location=device)

    if eval_mode and hasattr(model, "eval"):
        model.eval()

    return model


def load_model(config: ModelLoadConfig) -> Any:
    """Load a model using explicit or automatically inferred framework settings."""
    model_path = Path(config.model_path)
    ensure_model_exists(model_path)

    framework = config.framework
    if framework == "auto":
        framework = infer_framework(model_path)

    if framework == "keras":
        return load_keras_model(model_path, compile_model=config.compile_keras)

    if framework == "torch":
        return load_torch_model(
            model_path,
            device=config.device,
            eval_mode=config.eval_torch,
        )

    raise ValueError(f"Unsupported framework: {framework}")


def load_ct_model(
    model_path: str | Path = Path("models") / "ct_model.keras",
    framework: Framework = "auto",
    device: str = "cpu",
) -> Any:
    config = ModelLoadConfig(model_path=model_path, framework=framework, device=device)
    return load_model(config)


def load_xray_model(
    model_path: str | Path = Path("models") / "xray_model.keras",
    framework: Framework = "auto",
    device: str = "cpu",
) -> Any:
    config = ModelLoadConfig(model_path=model_path, framework=framework, device=device)
    return load_model(config)


def load_mri_model(
    model_path: str | Path = Path("models") / "mri_model.keras",
    framework: Framework = "auto",
    device: str = "cpu",
) -> Any:
    config = ModelLoadConfig(model_path=model_path, framework=framework, device=device)
    return load_model(config)


def load_all_models(
    ct_model_path: str | Path = Path("models") / "ct_model.keras",
    xray_model_path: str | Path = Path("models") / "xray_model.keras",
    mri_model_path: str | Path = Path("models") / "mri_model.keras",
    framework: Framework = "auto",
    device: str = "cpu",
) -> dict[str, Any]:
    # Pre-emptively create the models directory to avoid errors, although
    # the files themselves must still be provided by training or uploading.
    Path("models").mkdir(exist_ok=True)

    results = {}
    for name, path in [("ct", ct_model_path), ("xray", xray_model_path), ("mri", mri_model_path)]:
        try:
            results[name] = load_model(ModelLoadConfig(model_path=path, framework=framework, device=device))
        except FileNotFoundError:
            print(f"Warning: {name} model not found at {path}. Skipping.")
    return results


if __name__ == "__main__":
    models = load_all_models()
    if models:
        print(f"Loaded models: {', '.join(models.keys())}")
    else:
        print("No models were loaded. Please train your models or upload them to the 'models/' folder.")
```

```python
from __future__ import annotations
import csv
import os
from dataclasses import dataclass
from pathlib import Path
from typing import Any, List, Tuple, Dict, Optional
import tensorflow as tf

@dataclass
class TrainConfig:
    split_dir: Path
    model_output_path: Path
    history_output_path: Optional[Path] = None
    image_size: Tuple[int, int] = (224, 224)
    color_mode: str = "grayscale"
    batch_size: int = 32
    epochs: int = 20
    learning_rate: float = 0.001

@dataclass
class TrainResult:
    model: Any
    history: Any
    class_names: List[str]
    test_metrics: Optional[Dict[str, float]]
    model_path: Path

def channel_count(mode: str) -> int:
    return 1 if mode == "grayscale" else 3

def load_image_datasets(config: TrainConfig):
    train_ds = tf.keras.utils.image_dataset_from_directory(
        config.split_dir / "train",
        image_size=config.image_size,
        color_mode=config.color_mode,
        batch_size=config.batch_size,
        label_mode="int",
        shuffle=True,
    )
    val_ds = tf.keras.utils.image_dataset_from_directory(
        config.split_dir / "val",
        image_size=config.image_size,
        color_mode=config.color_mode,
        batch_size=config.batch_size,
        label_mode="int",
        shuffle=False,
    )
    test_ds = None
    if (config.split_dir / "test").exists():
        test_ds = tf.keras.utils.image_dataset_from_directory(
            config.split_dir / "test",
            image_size=config.image_size,
            color_mode=config.color_mode,
            batch_size=config.batch_size,
            label_mode="int",
            shuffle=False,
        )

    class_names = list(train_ds.class_names)
    autotune = tf.data.AUTOTUNE
    train_ds = train_ds.prefetch(autotune)
    val_ds = val_ds.prefetch(autotune)
    if test_ds is not None:
        test_ds = test_ds.prefetch(autotune)

    return train_ds, val_ds, test_ds, class_names

def build_cnn_model(image_size: tuple[int, int], color_mode: str, class_count: int, learning_rate: float) -> Any:
    layers = tf.keras.layers
    if class_count < 2: raise ValueError("At least two classes required.")

    input_shape = (*image_size, channel_count(color_mode))
    output_units = 1 if class_count == 2 else class_count
    output_activation = "sigmoid" if class_count == 2 else "softmax"
    loss = "binary_crossentropy" if class_count == 2 else "sparse_categorical_crossentropy"

    model = tf.keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Rescaling(1.0 / 255),
        layers.Conv2D(32, 3, activation="relu"),
        layers.MaxPooling2D(),
        layers.Conv2D(64, 3, activation="relu"),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128, activation="relu"),
        layers.Dense(output_units, activation=output_activation),
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), loss=loss, metrics=["accuracy"])
    return model

def write_history_csv(history: Any, path: Path):
    path.parent.mkdir(parents=True, exist_ok=True)
    data = history.history
    fieldnames = ["epoch"] + list(data.keys())
    with open(path, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        for i in range(len(next(iter(data.values())))):
            row = {"epoch": i + 1, **{k: v[i] for k, v in data.items()}}
            writer.writerow(row)

def train_image_model(config: TrainConfig) -> TrainResult:
    train_ds, val_ds, test_ds, class_names = load_image_datasets(config)
    model = build_cnn_model(config.image_size, config.color_mode, len(class_names), config.learning_rate)
    history = model.fit(train_ds, validation_data=val_ds, epochs=config.epochs)

    test_metrics = None
    if test_ds is not None:
        m = model.evaluate(test_ds, verbose=0)
        test_metrics = dict(zip(model.metrics_names, m))

    config.model_output_path.parent.mkdir(parents=True, exist_ok=True)
    model.save(config.model_output_path)
    if config.history_output_path: write_history_csv(history, config.history_output_path)

    return TrainResult(model, history, class_names, test_metrics, config.model_output_path)

def train_ct_model(split_dir: Path): return train_image_model(TrainConfig(split_dir, Path("models/ct_model.keras"), Path("models/ct_history.csv"), epochs=5))
def train_xray_model(split_dir: Path): return train_image_model(TrainConfig(split_dir, Path("models/xray_model.keras"), Path("models/xray_history.csv"), epochs=5))
def train_mri_model(split_dir: Path): return train_image_model(TrainConfig(split_dir, Path("models/mri_model.keras"), Path("models/mri_history.csv"), epochs=5))

if __name__ == "__main__":
    base = Path(os.environ.get("PULMO_VISION_DATASETS", "datasetsxctm")) / "splits"
    if (base / "ct").exists(): train_ct_model(base / "ct")
    if (base / "xray").exists(): train_xray_model(base / "xray")
    if (base / "mri").exists(): train_mri_model(base / "mri")
```

```python
from __future__ import annotations

import csv
import os
from dataclasses import dataclass
from pathlib import Path
from typing import Any, List, Optional, Tuple, Dict, Literal

import numpy as np
import tensorflow as tf


# Inlined from ydfXjxRhqDew
Framework = Literal["auto", "keras", "torch"]
KERAS_EXTENSIONS = {".h5", ".hdf5", ".keras"}
TORCH_EXTENSIONS = {".ckpt", ".pt", ".pth"}

def ensure_model_exists(model_path: Path) -> None:
    if not model_path.exists():
        raise FileNotFoundError(
            f"Model file or directory does not exist: {model_path}. "
            "Please ensure you have trained the models or uploaded them to this path."
        )

def load_keras_model(model_path: str | Path, compile_model: bool = False) -> Any:
    """Load a TensorFlow/Keras model."""
    model_path = Path(model_path)
    ensure_model_exists(model_path)

    try:
        from tensorflow import keras
    except ImportError as exc:
        raise ImportError(
            "TensorFlow is required to load Keras models. Install it with: "
            "pip install tensorflow"
        ) from exc

    return keras.models.load_model(model_path, compile=compile_model)

# Inlined from WPERcfNoqQyu
@dataclass
class TrainConfig:
    split_dir: Path
    model_output_path: Path
    history_output_path: Optional[Path] = None
    image_size: Tuple[int, int] = (224, 224)
    color_mode: str = "grayscale"
    batch_size: int = 32
    epochs: int = 20
    learning_rate: float = 0.001

def channel_count(mode: str) -> int:
    return 1 if mode == "grayscale" else 3

def load_image_datasets(config: TrainConfig):
    train_ds = tf.keras.utils.image_dataset_from_directory(
        config.split_dir / "train",
        image_size=config.image_size,
        color_mode=config.color_mode,
        batch_size=config.batch_size,
        label_mode="int",
        shuffle=True,
    )
    val_ds = tf.keras.utils.image_dataset_from_directory(
        config.split_dir / "val",
        image_size=config.image_size,
        color_mode=config.color_mode,
        batch_size=config.batch_size,
        label_mode="int",
        shuffle=False,
    )
    test_ds = None
    if (config.split_dir / "test").exists():
        test_ds = tf.keras.utils.image_dataset_from_directory(
            config.split_dir / "test",
            image_size=config.image_size,
            color_mode=config.color_mode,
            batch_size=config.batch_size,
            label_mode="int",
            shuffle=False,
        )

    class_names = list(train_ds.class_names)
    autotune = tf.data.AUTOTUNE
    train_ds = train_ds.prefetch(autotune)
    val_ds = val_ds.prefetch(autotune)
    if test_ds is not None:
        test_ds = test_ds.prefetch(autotune)

    return train_ds, val_ds, test_ds, class_names

def load_test_dataset(config: 'EvaluationConfig') -> Tuple[tf.data.Dataset, List[str]]:
    test_ds = tf.keras.utils.image_dataset_from_directory(
        config.split_dir / "test",
        image_size=config.image_size,
        color_mode=config.color_mode,
        batch_size=config.batch_size,
        label_mode="int",
        shuffle=False,
    )
    class_names = test_ds.class_names
    return test_ds.prefetch(tf.data.AUTOTUNE), class_names

def collect_true_labels(dataset: tf.data.Dataset) -> List[int]:
    true_labels = []
    for _, labels in dataset:
        true_labels.extend(labels.numpy())
    return true_labels

def predict_labels(model: Any, dataset: tf.data.Dataset, class_count: int) -> Tuple[List[int], List[float]]:
    predictions = model.predict(dataset)
    predicted_labels = tf.argmax(predictions, axis=1).numpy()
    confidences = tf.reduce_max(predictions, axis=1).numpy()
    return predicted_labels.tolist(), confidences.tolist()


@dataclass
class EvaluationConfig:
    split_dir: Path
    model_path: Path
    output_dir: Path
    image_size: Tuple[int, int] = (224, 224)
    color_mode: str = "grayscale"
    batch_size: int = 32


@dataclass
class EvaluationResult:
    metrics: dict[str, float]
    class_names: list[str]
    metrics_path: Path
    confusion_matrix_path: Path
    predictions_path: Path


def compute_confusion_matrix(
    true_labels: list[int],
    predicted_labels: list[int],
    class_count: int,
) -> list[list[int]]:
    confusion_matrix = np.zeros((class_count, class_count), dtype=int)
    for true, predicted in zip(true_labels, predicted_labels):
        confusion_matrix[true][predicted] += 1
    return confusion_matrix.tolist()


def compute_metrics(
    true_labels: list[int],
    predicted_labels: list[int],
    confusion_matrix: list[list[int]],
) -> dict[str, float]:
    metrics = {}
    class_count = len(confusion_matrix)

    per_class_precision = []
    per_class_recall = []
    per_class_f1 = []

    for class_id in range(class_count):
        true_positive = confusion_matrix[class_id][class_id]
        false_positive = sum(
            confusion_matrix[j][class_id]
            for j in range(class_count)
            if j != class_id
        )
        false_negative = sum(
            confusion_matrix[class_id][j]
            for j in range(class_count)
            if j != class_id
        )

        precision = (
            true_positive / (true_positive + false_positive)
            if (true_positive + false_positive) > 0
            else 0.0
        )
        recall = (
            true_positive / (true_positive + false_negative)
            if (true_positive + false_negative) > 0
            else 0.0
        )
        f1_score = (
            2 * precision * recall / (precision + recall)
            if (precision + recall) > 0
            else 0.0
        )

        per_class_precision.append(precision)
        per_class_recall.append(recall)
        per_class_f1.append(f1_score)

    metrics["macro_precision"] = (
        sum(per_class_precision) / class_count if class_count > 0 else 0.0
    )
    metrics["macro_recall"] = (
        sum(per_class_recall) / class_count if class_count > 0 else 0.0
    )
    metrics["macro_f1"] = sum(per_class_f1) / class_count if class_count > 0 else 0.0

    return metrics

def write_metrics_csv(metrics_path: Path, metrics: dict[str, float]) -> Path:
    metrics_path.parent.mkdir(parents=True, exist_ok=True)

    with metrics_path.open("w", newline="", encoding="utf-8") as csv_file:
        writer = csv.DictWriter(csv_file, fieldnames=["metric", "value"])
        writer.writeheader()
        for metric, value in metrics.items():
            writer.writerow({"metric": metric, "value": value})

    return metrics_path

def write_confusion_matrix_csv(
    confusion_matrix_path: Path,
    confusion_matrix: list[list[int]],
    class_names: list[str],
) -> Path:
    confusion_matrix_path.parent.mkdir(parents=True, exist_ok=True)

    with confusion_matrix_path.open("w", newline="", encoding="utf-8") as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(["actual/predicted", *class_names])

        for class_name, row in zip(class_names, confusion_matrix, strict=True):
            writer.writerow([class_name, *row])

    return confusion_matrix_path

def write_predictions_csv(
    predictions_path: Path,
    true_labels: list[int],
    predicted_labels: list[int],
    confidences: list[float],
    class_names: list[str],
) -> Path:
    predictions_path.parent.mkdir(parents=True, exist_ok=True)

    with predictions_path.open("w", newline="", encoding="utf-8") as csv_file:
        writer = csv.DictWriter(
            csv_file,
            fieldnames=[
                "sample_index",
                "actual_label",
                "actual_class",
                "predicted_label",
                "predicted_class",
                "confidence",
            ],
        )
        writer.writeheader()

        for index, (true_label, predicted_label, confidence) in enumerate(
            zip(true_labels, predicted_labels, confidences, strict=True),
            start=1,
        ):
            writer.writerow(
                {
                    "sample_index": index,
                    "actual_label": true_label,
                    "actual_class": class_names[true_label],
                    "predicted_label": predicted_label,
                    "predicted_class": class_names[predicted_label],
                    "confidence": confidence,
                }
            )

    return predictions_path

def evaluate_image_model(config: EvaluationConfig) -> EvaluationResult:
    output_dir = Path(config.output_dir)
    model = load_keras_model(config.model_path)
    test_dataset, class_names = load_test_dataset(config)

    true_labels = collect_true_labels(test_dataset)
    predicted_labels, confidences = predict_labels(model, test_dataset, len(class_names))
    confusion_matrix = compute_confusion_matrix(
        true_labels,
        predicted_labels,
        len(class_names),
    )
    metrics = compute_metrics(true_labels, predicted_labels, confusion_matrix)

    metrics_path = write_metrics_csv(output_dir / "metrics.csv", metrics)
    confusion_matrix_path = write_confusion_matrix_csv(
        output_dir / "confusion_matrix.csv",
        confusion_matrix,
        class_names,
    )
    predictions_path = write_predictions_csv(
        output_dir / "predictions.csv",
        true_labels,
        predicted_labels,
        confidences,
        class_names,
    )

    return EvaluationResult(
        metrics=metrics,
        class_names=class_names,
        metrics_path=metrics_path,
        confusion_matrix_path=confusion_matrix_path,
        predictions_path=predictions_path,
    )

def evaluate_ct_model(
    split_dir: str | Path,
    model_path: str | Path = Path("models") / "ct_model.keras",
    output_dir: str | Path = Path("models") / "ct_evaluation",
) -> EvaluationResult:
    config = EvaluationConfig(
        split_dir=split_dir,
        model_path=model_path,
        output_dir=output_dir,
    )
    return evaluate_image_model(config)

def evaluate_xray_model(
    split_dir: str | Path,
    model_path: str | Path = Path("models") / "xray_model.keras",
    output_dir: str | Path = Path("models") / "xray_evaluation",
) -> EvaluationResult:
    config = EvaluationConfig(
        split_dir=split_dir,
        model_path=model_path,
        output_dir=output_dir,
    )
    return evaluate_image_model(config)

def evaluate_mri_model(
    split_dir: str | Path,
    model_path: str | Path = Path("models") / "mri_model.keras",
    output_dir: str | Path = Path("models") / "mri_evaluation",
) -> EvaluationResult:
    config = EvaluationConfig(
        split_dir=split_dir,
        model_path=model_path,
        output_dir=output_dir,
    )
    return evaluate_image_model(config)

if __name__ == "__main__":
    # Use the same base directory as the download and split scripts
    split_base_dir = Path(os.environ.get("PULMO_VISION_DATASETS", "datasetsxctm")) / "splits"

    evaluate_ct_model(split_base_dir / "ct")
    evaluate_xray_model(split_base_dir / "xray")
    evaluate_mri_model(split_base_dir / "mri")
```

```python
from __future__ import annotations

import csv
import os
from dataclasses import dataclass
from pathlib import Path
from typing import Any, Iterable, List, Optional, Tuple, Literal

from PIL import Image, ImageOps # Added
import numpy as np # Added
import tensorflow as tf # Added
from tqdm.auto import tqdm # Added for progress bar

# Dataclass definitions for testing and prediction results
@dataclass(frozen=True)
class TestConfig:
    model_path: str | Path
    input_path: str | Path
    output_csv_path: str | Path
    image_size: Tuple[int, int] = (224, 224)
    color_mode: str = "grayscale"
    class_names: List[str] | None = None

@dataclass(frozen=True)
class PredictionResult:
    image_path: Path
    predicted_label: int
    predicted_class: str
    confidence: float
    status: str = "ok"

# Helper functions for image loading and path collection
def iter_files(path: Path):
    for root, _, files in os.walk(path):
        for file in files:
            yield Path(root) / file

def is_image_file(path: Path) -> bool:
    return path.suffix.lower() in [".jpg", ".jpeg", ".png", ".gif", ".bmp", ".tiff", ".webp"]

def collect_image_paths(input_path: str | Path) -> list[Path]:
    input_path = Path(input_path)
    if not input_path.exists():
        raise FileNotFoundError(f"Input path does not exist: {input_path}")
    return [p for p in iter_files(input_path) if is_image_file(p)]

def load_image_batch(
    path: Path,
    image_size: tuple[int, int],
    color_mode: str,
) -> tf.Tensor:
    img = Image.open(path)
    img = ImageOps.exif_transpose(img) # Apply exif orientation if any
    img = ImageOps.fit(img, image_size, Image.Resampling.LANCZOS)
    if color_mode == "grayscale":
        img = img.convert("L")
    elif color_mode == "rgb":
        img = img.convert("RGB")
    else:
        raise ValueError(f"Unsupported color_mode: {color_mode}")
    img_array = tf.keras.utils.img_to_array(img)
    return tf.expand_dims(img_array, 0)  # Create a batch

# Helper function to load Keras model (inlined from BkEjCOeE33Jt for self-containment)
def ensure_model_exists(model_path: Path) -> None:
    if not model_path.exists():
        raise FileNotFoundError(
            f"Model file or directory does not exist: {model_path}. "
            "Please ensure you have trained the models or uploaded them to this path."
        )

def load_keras_model(model_path: str | Path, compile_model: bool = False) -> Any:
    """Load a TensorFlow/Keras model."""
    model_path = Path(model_path)
    ensure_model_exists(model_path)

    try:
        from tensorflow import keras
    except ImportError as exc:
        raise ImportError(
            "TensorFlow is required to load Keras models. Install it with: "
            "pip install tensorflow"
        ) from exc

    return keras.models.load_model(model_path, compile=compile_model)

def default_class_names(class_count: int) -> list[str]:
    return [f"class_{index}" for index in range(class_count)]

def normalize_prediction_values(prediction: Any) -> list[float]:
    values = prediction.tolist()
    if isinstance(values, list):
        return [float(value) for value in values]
    return [float(values)]

def decode_prediction(
    prediction: Any,
    class_names: list[str] | None = None,
) -> tuple[int, str, float]:
    values = normalize_prediction_values(prediction)

    if len(values) == 1:
        probability = values[0]
        predicted_label = 1 if probability >= 0.5 else 0
        names = class_names or default_class_names(2)
        confidence = probability if predicted_label == 1 else 1.0 - probability
        return predicted_label, names[predicted_label], confidence

    predicted_label = max(range(len(values)), key=values.__getitem__)
    names = class_names or default_class_names(len(values))
    return predicted_label, names[predicted_label], values[predicted_label]

def predict_single_image(
    model: Any,
    image_path: str | Path,
    image_size: tuple[int, int] = (224, 224),
    color_mode: str = "grayscale",
    class_names: list[str] | None = None,
) -> PredictionResult:
    image_path = Path(image_path)
    image_batch = load_image_batch(image_path, image_size, color_mode)
    prediction = model.predict(image_batch, verbose=0)[0]
    predicted_label, predicted_class, confidence = decode_prediction(
        prediction,
        class_names=class_names,
    )

    return PredictionResult(
        image_path=image_path,
        predicted_label=predicted_label,
        predicted_class=predicted_class,
        confidence=confidence,
    )

def test_model(config: TestConfig) -> list[PredictionResult]:
    model = load_keras_model(config.model_path)
    image_paths = collect_image_paths(config.input_path)

    # Wrap image_paths with tqdm for a progress bar
    return [
        predict_single_image(
            model=model,
            image_path=image_path,
            image_size=config.image_size,
            color_mode=config.color_mode,
            class_names=config.class_names,
        )
        for image_path in tqdm(image_paths, desc="Predicting images")
    ]

def write_predictions_csv(
    output_csv_path: str | Path,
    predictions: Iterable[PredictionResult],
) -> Path:
    output_csv_path = Path(output_csv_path)
    output_csv_path.parent.mkdir(parents=True, exist_ok=True)

    with output_csv_path.open("w", newline="", encoding="utf-8") as csv_file:
        writer = csv.DictWriter(
            csv_file,
            fieldnames=[
                "image_path",
                "predicted_label",
                "predicted_class",
                "confidence",
                "status", # Added status to CSV header
            ],
        )
        writer.writeheader()

        for prediction in predictions:
            writer.writerow(
                {
                    "image_path": prediction.image_path,
                    "predicted_label": prediction.predicted_label,
                    "predicted_class": prediction.predicted_class,
                    "confidence": prediction.confidence,
                    "status": prediction.status, # Added status to CSV row
                }
            )

    return output_csv_path

def test_and_save_predictions(config: TestConfig) -> list[PredictionResult]:
    predictions = test_model(config)
    write_predictions_csv(config.output_csv_path, predictions)
    return predictions

def test_ct_model(
    input_path: str | Path,
    model_path: str | Path = Path("models") / "ct_model.keras",
    output_csv_path: str | Path = Path("models") / "ct_test_predictions.csv",
    class_names: list[str] | None = None,
) -> list[PredictionResult]:
    config = TestConfig(
        model_path=model_path,
        input_path=input_path,
        output_csv_path=output_csv_path,
        class_names=class_names,
    )
    return test_and_save_predictions(config)

def test_xray_model(
    input_path: str | Path,
    model_path: str | Path = Path("models") / "xray_model.keras",
    output_csv_path: str | Path = Path("models") / "xray_test_predictions.csv",
    class_names: list[str] | None = None,
) -> list[PredictionResult]:
    config = TestConfig(
        model_path=model_path,
        input_path=input_path,
        output_csv_path=output_csv_path,
        class_names=class_names,
    )
    return test_and_save_predictions(config)

def test_mri_model(
    input_path: str | Path,
    model_path: str | Path = Path("models") / "mri_model.keras",
    output_csv_path: str | Path = Path("models") / "mri_test_predictions.csv",
    class_names: list[str] | None = None,
) -> list[PredictionResult]:
    config = TestConfig(
        model_path=model_path,
        input_path=input_path,
        output_csv_path=output_csv_path,
        class_names=class_names,
    )
    return test_and_save_predictions(config)

if __name__ == "__main__":
    # Use the base directory for datasets, and specifically the 'test' splits.
    base_dataset_dir = Path(os.environ.get("PULMO_VISION_DATASETS", "datasetsxctm"))
    sample_dir = base_dataset_dir / "splits"

    test_ct_model(sample_dir / "ct" / "test")
    test_xray_model(sample_dir / "xray" / "test")
    test_mri_model(sample_dir / "mri" / "test")
```

```python
from __future__ import annotations

import csv
import os
from dataclasses import dataclass
from pathlib import Path
from typing import Any, Iterable, List, Literal, Optional, Tuple

from PIL import Image, ImageOps
import numpy as np
import tensorflow as tf
from tqdm.auto import tqdm


# Define DatasetType as it's used in PredictionConfig
DatasetType = Literal["ct", "xray", "mri"]


@dataclass(frozen=True)
class PredictionConfig:
    dataset_type: DatasetType
    image_path: str | Path
    model_path: str | Path | None = None
    image_size: Tuple[int, int] = (224, 224)
    color_mode: str = "grayscale"
    class_names: List[str] | None = None


@dataclass(frozen=True)
class Prediction:
    dataset_type: DatasetType
    image_path: Path
    predicted_label: int
    predicted_class: str
    confidence: float
    probabilities: dict[str, float]


# Inlined helper functions from VoF4sxSViGNY and potentially others
def iter_files(path: Path):
    for root, _, files in os.walk(path):
        for file in files:
            yield Path(root) / file

def is_image_file(path: Path) -> bool:
    return path.suffix.lower() in [".jpg", ".jpeg", ".png", ".gif", ".bmp", ".tiff", ".webp"]

def collect_image_paths(input_path: str | Path) -> list[Path]:
    input_path = Path(input_path)

    if not input_path.exists():
        raise FileNotFoundError(f"Input path does not exist: {input_path}")

    if input_path.is_file():
        if not is_image_file(input_path):
            raise ValueError(f"Unsupported file type: {input_path}")
        return [input_path]

    return [p for p in iter_files(input_path) if is_image_file(p)]

def load_image_for_prediction(
    image_path: Path,
    image_size: tuple[int, int],
    color_mode: str,
) -> tf.Tensor:
    img = Image.open(image_path)
    img = ImageOps.exif_transpose(img)
    img = ImageOps.fit(img, image_size, Image.Resampling.LANCZOS)
    if color_mode == "grayscale":
        img = img.convert("L")
    elif color_mode == "rgb":
        img = img.convert("RGB")
    else:
        raise ValueError(f"Unsupported color_mode: {color_mode}")
    img_array = tf.keras.utils.img_to_array(img)
    return tf.expand_dims(img_array, 0)

def ensure_model_exists(model_path: Path) -> None:
    if not model_path.exists():
        raise FileNotFoundError(
            f"Model file or directory does not exist: {model_path}. "
            "Please ensure you have trained the models or uploaded them to this path."
        )

def load_keras_model(model_path: str | Path) -> Any:
    """Load a TensorFlow/Keras model."""
    model_path = Path(model_path)
    ensure_model_exists(model_path)
    # tensorflow import is already at the top of the cell
    return tf.keras.models.load_model(model_path, compile=False) # Assuming compile=False is fine for prediction

def default_class_names(class_count: int) -> list[str]:
    return [f"class_{index}" for index in range(class_count)]

def normalize_prediction_values(prediction: Any) -> list[float]:
    values = prediction.tolist()
    if isinstance(values, list):
        return [float(value) for value in values]
    return [float(values)]

def resolve_model_path(
    dataset_type: DatasetType, model_path: str | Path | None
) -> Path:
    if model_path:
        return Path(model_path)
    return Path("models") / f"{dataset_type}_model.keras"

def decode_probabilities(values: Any, class_names: list[str] | None = None) -> tuple[int, str, float, dict[str, float]]:
    names = class_names or default_class_names(len(values))

    if len(names) != len(values):
        raise ValueError(
            f"class_names length ({len(names)}) must match prediction length ({len(values)})."
        )

    predicted_label = max(range(len(values)), key=values.__getitem__)
    predicted_class = names[predicted_label]
    probabilities = dict(zip(names, values, strict=True))

    return predicted_label, predicted_class, values[predicted_label], probabilities


def predict_with_model(
    model: Any,
    config: PredictionConfig,
) -> Prediction:
    image_path = Path(config.image_path)
    image_batch = load_image_for_prediction(
        image_path=image_path,
        image_size=config.image_size,
        color_mode=config.color_mode,
    )
    raw_prediction = model.predict(image_batch, verbose=0)[0]
    predicted_label, predicted_class, confidence, probabilities = decode_probabilities(
        raw_prediction,
        class_names=config.class_names,
    )

    return Prediction(
        dataset_type=config.dataset_type,
        image_path=image_path,
        predicted_label=predicted_label,
        predicted_class=predicted_class,
        confidence=confidence,
        probabilities=probabilities,
    )


def predict_image(config: PredictionConfig) -> Prediction:
    model_path = resolve_model_path(config.dataset_type, config.model_path)
    model = load_keras_model(model_path)
    return predict_with_model(model, config)


def predict_folder(
    dataset_type: DatasetType,
    input_dir: str | Path,
    model_path: str | Path | None = None,
    image_size: tuple[int, int] = (224, 224),
    color_mode: str = "grayscale",
    class_names: list[str] | None = None,
) -> list[Prediction]:
    resolved_model_path = resolve_model_path(dataset_type, model_path)
    model = load_keras_model(resolved_model_path)
    image_paths = collect_image_paths(input_dir)

    return [
        predict_with_model(
            model,
            PredictionConfig(
                dataset_type=dataset_type,
                image_path=image_path,
                model_path=resolved_model_path,
                image_size=image_size,
                color_mode=color_mode,
                class_names=class_names,
            ),
        )
        for image_path in image_paths
    ]


def write_predictions_csv(
    predictions: Iterable[Prediction],
    output_csv_path: str | Path,
) -> Path:
    output_csv_path = Path(output_csv_path)
    output_csv_path.parent.mkdir(parents=True, exist_ok=True)

    with output_csv_path.open("w", newline="", encoding="utf-8") as csv_file:
        writer = csv.DictWriter(
            csv_file,
            fieldnames=[
                "dataset_type",
                "image_path",
                "predicted_label",
                "predicted_class",
                "confidence",
                "probabilities",
            ],
        )
        writer.writeheader()

        for prediction in predictions:
            writer.writerow(
                {
                    "dataset_type": prediction.dataset_type,
                    "image_path": prediction.image_path,
                    "predicted_label": prediction.predicted_label,
                    "predicted_class": prediction.predicted_class,
                    "confidence": prediction.confidence,
                    "probabilities": prediction.probabilities,
                }
            )

    return output_csv_path


def predict_ct_image(
    image_path: str | Path,
    model_path: str | Path | None = None,
    class_names: list[str] | None = None,
) -> Prediction:
    config = PredictionConfig(
        dataset_type="ct",
        image_path=image_path,
        model_path=model_path,
        class_names=class_names,
    )
    return predict_image(config)


def predict_xray_image(
    image_path: str | Path,
    model_path: str | Path | None = None,
    class_names: list[str] | None = None,
) -> Prediction:
    config = PredictionConfig(
        dataset_type="xray",
        image_path=image_path,
        model_path=model_path,
        class_names=class_names,
    )
    return predict_image(config)


def predict_mri_image(
    image_path: str | Path,
    model_path: str | Path | None = None,
    class_names: list[str] | None = None,
) -> Prediction:
    config = PredictionConfig(
        dataset_type="mri",
        image_path=image_path,
        model_path=model_path,
        class_names=class_names,
    )
    return predict_image(config)


if __name__ == "__main__":
    # The hardcoded path below caused a FileNotFoundError as it's Windows-specific.
    # We now dynamically find a sample image from the downloaded datasets.
    base_data_dir = Path(os.environ.get("PULMO_VISION_DATASETS", "datasetsxctm"))
    ct_dataset_dir = base_data_dir / "ct"

    sample_image = None
    try:
        # Collect all image paths from the downloaded CT datasets
        all_ct_images = collect_image_paths(ct_dataset_dir)
        if all_ct_images:
            sample_image = all_ct_images[0] # Take the first image found
        else:
            print(f"Error: No CT images found in {ct_dataset_dir}. Please ensure datasets were downloaded correctly.")
    except FileNotFoundError:
        print(f"Error: CT dataset directory not found at {ct_dataset_dir}. Please run the download script first.")
    except Exception as e:
        print(f"An error occurred while trying to find a sample CT image: {e}")

    if sample_image:
        try:
            prediction = predict_ct_image(sample_image)
            print(
                f"{prediction.image_path}: "
                f"{prediction.predicted_class} ({prediction.confidence:.4f})"
            )
        except FileNotFoundError as e:
            print(f"Error: {e}. Please ensure the sample image path is correct and accessible.")
        except Exception as e:
            print(f"An unexpected error occurred during prediction: {e}")
```
